```python
# ========================================
# 🎯 Phân tích sự phân bố học lực (Grade) theo hoạt động ngoại khóa (Extracurricular Activities)
# ========================================

# Tính tỷ lệ phần trăm (%) các loại học lực trong mỗi nhóm học sinh:
# - Nhóm 1: Có tham gia hoạt động ngoại khóa (Yes)
# - Nhóm 2: Không tham gia (No)
# Dữ liệu đầu vào: df — bảng dữ liệu sinh viên đã xử lý sạch

extra_grade_dist = pd.crosstab(
    df['Extracurricular_Activities'],  # Chỉ số hàng: Có hoặc không tham gia ngoại khóa
    df['Grade'],                       # Chỉ số cột: Các mức Grade (A, B, C, D)
    normalize='index'                          # Chuẩn hóa theo từng hàng (tính % trong từng nhóm)
) * 100  # Nhân 100 để chuyển thành phần trăm (%)


#  Vẽ biểu đồ cột nhóm (grouped bar chart) để so sánh trực quan

extra_grade_dist.plot(
    kind='bar',                # Kiểu biểu đồ cột theo nhóm
    figsize=(12, 6),           # Kích thước biểu đồ (chiều rộng x chiều cao)
    colormap='viridis'         # Bảng màu đẹp mắt
)

# Thiết lập nhãn và tiêu đề cho biểu đồ
plt.title('Phân bố điểm theo hoạt động ngoại khóa', fontsize=14)
plt.xlabel('Extracurricular Activities', fontsize=12)
plt.ylabel('Percentage of Students', fontsize=12)
plt.legend(title='Grade')  # Thêm tiêu đề cho phần chú giải

# Bước 4️⃣: Thêm lưới trục y để dễ đọc tỷ lệ
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Tự động căn chỉnh bố cục biểu đồ cho đẹp
plt.tight_layout()

# Hiển thị biểu đồ
plt.show()

# In bảng số liệu phần trăm phân phối (có thể dùng để tạo báo cáo hoặc phân tích sâu)
extra_grade_dist
```

```python
# Thiết lập seaborn cho đẹp hơn
sns.set(style="whitegrid")

# 1. Lấy mẫu đều từ mỗi khoa
sample_size = 1000  # Số lượng mẫu muốn lấy từ mỗi khoa
df_sampled = df.groupby("Department").apply(lambda x: x.sample(n=sample_size, random_state=2500)).reset_index(drop=True)

# 2. Tính điểm trung bình theo khoa
dept_avg = df_sampled.groupby("Department")["Total_Score"].mean()

# 3. Tính điểm danh trung bình theo khoa
attendance_avg = df_sampled.groupby("Department")["Attendance (%)"].mean()

# 4. Đếm số lượng học sinh theo từng loại điểm (Grade) trong mỗi khoa
dept_grade_cnt = pd.crosstab(df_sampled['Department'], df_sampled['Grade'])

# 5. Tạo biểu đồ gồm 3 hình con
fig, axes = plt.subplots(1, 2, figsize=(16, 9))  # 1 hàng, 3 cột

# 5.1 Biểu đồ Pie thể hiện phân phối điểm trung bình theo khoa
axes[0].pie(
    dept_avg,                                 # Giá trị điểm trung bình
    labels=dept_avg.index,                    # Nhãn là tên khoa
    autopct='%1.1f%%',                        # Hiển thị phần trăm
    colors=sns.color_palette("mako", len(dept_avg))  # Màu
)
axes[0].set_title("Equal Sampled Total Score Distribution", fontsize=14, fontweight="bold")


# 5.3 Biểu đồ cột xếp chồng thể hiện phân phối điểm (Grade) theo khoa
bar_colors = sns.color_palette("Set2", n_colors=len(dept_grade_cnt.columns))  # Màu cho từng loại Grade
dept_grade_cnt.plot(
    kind="bar",              # Kiểu biểu đồ cột
    stacked=True,            # Cột xếp chồng
    color=bar_colors,        
    ax=axes[1],              # Trục thứ 3
    edgecolor="black",       # Viền cột
    linewidth=1.2            # Độ dày viền
)
axes[1].set_title("Grade Distribution Across Departments", fontsize=14, fontweight="bold")
axes[1].set_xlabel("Department")
axes[1].set_ylabel("Number of Students")
axes[1].tick_params(axis='x', rotation=0)  # Không xoay trục X

# 6. Tối ưu bố cục và hiển thị
plt.tight_layout()
plt.show()
```

# Giới hạn của phân tích tuyến tính (Linear Correlation)

- Biểu đồ nhiệt gửi chỉ dựa trên tương quan tuyến tính (linear correlation) — đo mối quan hệ theo đường thẳng giữa 2 biến. Tuy nhiên:

- Nếu mối quan hệ giữa hai yếu tố không phải tuyến tính, thì hệ số tương quan gần bằng 0, **nhưng không có nghĩa là không có mối quan hệ**.

--- Ví dụ: một mối quan hệ hình parabol (∩ hoặc ∪) sẽ có hệ số tương quan gần 0, dù hai biến vẫn ảnh hưởng nhau mạnh.


# 2. Các mô hình phi tuyến giải quyết điều này thế nào?
## ✅ Random Forest
Là mô hình ensemble tree-based (tập hợp nhiều cây quyết định).

Nó không cần mối quan hệ tuyến tính, mà học theo cách chia nhỏ không gian dữ liệu thành các điều kiện logic.

Ưu điểm:

Xử lý tốt mối quan hệ phức tạp giữa các biến.

Có thể đo mức độ quan trọng (feature importance) của từng yếu tố ảnh hưởng đến kết quả.

## ✅ XGBoost (Extreme Gradient Boosting)
Là một mô hình Boosting Tree mạnh mẽ, học theo cách tuần tự cải thiện lỗi của mô hình trước.

Khả năng mô hình hóa tốt các tương tác phức tạp giữa nhiều yếu tố.

Dễ tùy biến và hiệu quả cao trong dự đoán điểm số hoặc phân loại học lực.

- 🔍 Ví dụ về mối quan hệ phi tuyến:
Giả sử có mối quan hệ như sau:

Nếu học sinh ngủ quá ít (<4h) hoặc ngủ quá nhiều (>10h) thì điểm kém → mối quan hệ dạng parabol (U ngược).
